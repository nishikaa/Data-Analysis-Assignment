# Fetch Data Exploratory Analysis- Data Quality And Structure

## üìä Analysis Summary

This notebook provides an initial exploration of the provided datasets: `USER_TAKEHOME.csv`, `TRANSACTION_TAKEHOME.csv`, and `PRODUCTS_TAKEHOME.csv`.  
The goal is to identify data quality issues and understand the structure and relationships within the data.

---

### üë§ User Data (`USER_TAKEHOME.csv`)

- This dataset contains **100,000 rows** and **6 columns**:  
  `'ID'`, `'CREATED_DATE'`, `'BIRTH_DATE'`, `'STATE'`, `'LANGUAGE'`, and `'GENDER'`.
- All columns except `'LANGUAGE'` have over **94% non-null coverage**.  
  `'ID'` and `'CREATED_DATE'` are **100% complete**.
- **No duplicate rows** were found, suggesting that `'ID'` can be treated as a **primary key**.
- This is a **dimensional dataset** representing user-level data.
- All fields are of `object` data type; `'CREATED_DATE'` and `'BIRTH_DATE'` should be converted to proper `datetime` types for analysis.

---

### üßæ Transaction Data (`TRANSACTION_TAKEHOME.csv`)

- This dataset contains **50,000 rows** and **8 columns**:  
  `'RECEIPT_ID'`, `'PURCHASE_DATE'`, `'SCAN_DATE'`, `'STORE_NAME'`, `'USER_ID'`, `'BARCODE'`, `'FINAL_QUANTITY'`, and `'FINAL_SALE'`.
- All columns except `'BARCODE'` are nearly 100% complete; `'BARCODE'` has approximately **90% coverage**.
- The dataset contains **171 duplicate rows**, which should be removed before further analysis.
- This is a **fact table**, where:
  - `'USER_ID'` can join to the User dataset
  - `'BARCODE'` can join to the Product dataset
- Most fields are `object` type. `'BARCODE'` is stored as `float` and should be appropriately converted for accurate joins.

---

### üè∑Ô∏è Product Data (`PRODUCTS_TAKEHOME.csv`)

- This dataset contains **845,552 rows** and **7 columns**:  
  `'CATEGORY_1'`, `'CATEGORY_2'`, `'CATEGORY_3'`, `'CATEGORY_4'`, `'MANUFACTURER'`, `'BRAND'`, and `'BARCODE'`.
- All columns contain some null values. However:
  - `'CATEGORY_1'`, `'CATEGORY_2'`, and `'BARCODE'` have **high coverage**
  - `'CATEGORY_4'` has **very poor coverage (< 20%)** and is likely **not useful for analysis**
- There are **215 duplicate rows** that should be removed.
- This is a **dimensional dataset**, with `'BARCODE'` acting as a **foreign key** for joins with the Transaction data.
- All fields are `object` type except `'BARCODE'`, which is `float` and should be converted to string or integer for matching.

---

## ‚ùì Questions To Address

### 1. Are there any data quality issues present?
Yes ‚Äî the main issues observed across datasets include:
- **Null values** (especially in `'LANGUAGE'` and `'CATEGORY_4'`)
- **Duplicate rows** in both Transaction and Product datasets
- **Inconsistent or inappropriate data types**, such as dates stored as strings and numeric IDs as floats

These issues have been noted and addressed during data exploration.

### 2. Are there any fields that are challenging to understand?
Overall, most columns are self-explanatory.  
The only ambiguity lies in the **`CATEGORY_1` to `CATEGORY_4`** columns in the product dataset:
- These appear to represent a **hierarchy of product categorization**
- From exploration, it seems:
  - `'CATEGORY_1'` provides a **broad** product classification
  - `'CATEGORY_4'` offers a **very specific** description  
  - The categories become more granular as they progress from 1 to 4

### 3. Missing Values Check

```python
user_df.isnull().sum()
```

```python
missing_pct = user_df.isnull().mean().sort_values(ascending=False) * 100
missing_pct.plot(kind='barh', color='coral', figsize=(8, 5))
plt.title("Percentage of Missing Values by Column-User")
plt.xlabel("Percent Missing")
plt.tight_layout()
plt.show()
```

```python
print("Missing Values in TRANSACTION_TAKEHOME:")
transaction_df.isnull().sum()
```

```python
missing_pct = transaction_df.isnull().mean().sort_values(ascending=False) * 100
missing_pct.plot(kind='barh', color='coral', figsize=(8, 5))
plt.title("Percentage of Missing Values by Column-Product")
plt.xlabel("Percent Missing")
plt.tight_layout()
plt.show()
```

```python
print("Missing Values in PRODUCTS_TAKEHOME:")
products_df.isnull().sum()
```

```python
missing_pct = products_df.isnull().mean().sort_values(ascending=False) * 100
missing_pct.plot(kind='barh', color='coral', figsize=(8, 5))
plt.title("Percentage of Missing Values by Column-Product")
plt.xlabel("Percent Missing")
plt.tight_layout()
plt.show()
```
